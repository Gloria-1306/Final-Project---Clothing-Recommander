Traiter les NaN : Stratégie de remplacement des valeurs manquantes :
Puisque les données sont liées aux pays, tu peux remplacer les NaN en prenant la médiane ou la moyenne des pays ayant un IDH (Human Development Index) similaire.

Étapes à suivre :
Regrouper les pays par catégorie d'IDH
(ex. : faible, moyen, élevé, très élevé)
Calculer la médiane ou la moyenne du groupe
Remplacer les NaN par cette valeur

```python

# 4️⃣ Remplir les NaN par la médiane des pays ayant un IDH similaire
# Définir des groupes d'IDH
hdi_copy['hdi_group'] = pd.qcut(hdi_copy['hdi_value'], q=5, labels=False)

# Appliquer cette classification aux autres df
for df in [clean_degraded_land, clean_indus_water, clean_total_fatal]:
    df = df.merge(hdi_copy[['country', 'hdi_group']], on='country', how='left')

    # Remplacer les NaN par la médiane du groupe IDH
    for col in df.select_dtypes(include=['float', 'int']).columns:  # Seulement les colonnes numériques
        df[col] = df.groupby('hdi_group')[col].transform(lambda x: x.fillna(x.median()))

print("🚀 Données harmonisées et NaN remplis avec la médiane des pays ayant un IDH similaire !")
```

Pourquoi l'IQR et pas le Z-score ?
Le Z-score suppose une distribution normale des données, ce qui est rarement le cas avec des données environnementales (qui ont souvent une longue traîne).
L'IQR est plus robuste aux valeurs extrêmes et fonctionne bien avec des données asymétriques comme les émissions de gaz à effet de serre.

```python
# Suppression des lignes où toutes les colonnes de 1990 à 2021 sont nulles ou nulles
years = [str(year) for year in range(1990, 2022)]  # Liste des colonnes années
df = df_industrial_emissions_filtered.loc[~(df[years] == 0).all(axis=1)]  # Supprime les pays avec toutes les valeurs à 0

# Calcul des quartiles et IQR pour identifier les outliers
Q1 = df[years].quantile(0.25)
Q3 = df[years].quantile(0.75)
IQR = Q3 - Q1

# Définition des bornes pour éliminer les outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrer les valeurs aberrantes
df_filtered= df[~((df[years] < lower_bound) | (df[years] > upper_bound)).any(axis=1)]

# Vérification des résultats
print(f"Nombre de lignes avant filtrage : {df.shape[0]}")
print(f"Nombre de lignes après filtrage : {df_filtered.shape[0]}")
```
