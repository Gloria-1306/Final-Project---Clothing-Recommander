```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv("/kaggle/input/vestiaire-fashion-dataset/vestiaire.csv")

# Convert appropriate columns to categorical
df['product_gender_target'] = df['product_gender_target'].astype('category')
df['product_condition'] = df['product_condition'].astype('category')
df['seller_country'] = df['seller_country'].astype('category')

# Set a consistent style and color palette for plots
sns.set(style="whitegrid")
black_palette = ["#000000", "#333333", "#666666", "#999999"]
```

```python
# 1. Distribution of Product Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price_usd'], bins=50, kde=True, color=black_palette[0])
plt.title('Distribution of Product Prices', fontsize=16)
plt.xlabel('Price in USD')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Top Brands by Number of Listings
top_brands = df['brand_name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_brands.values, y=top_brands.index, palette=black_palette)
plt.title('Top Brands by Number of Listings', fontsize=16)
plt.xlabel('Number of Listings')
plt.ylabel('Brand Name')
plt.show()
```

```python
# 3. Distribution of Product Conditions
plt.figure(figsize=(10, 6))
sns.countplot(y='product_condition', data=df, palette=black_palette)
plt.title('Distribution of Product Conditions', fontsize=16)
plt.xlabel('Frequency')
plt.ylabel('Product Condition')
plt.show()
```

```python
# 4. Geographical Distribution of Sellers
top_countries = df['seller_country'].value_counts().head(10)
plt.figure(figsize=(10, 25))
sns.barplot(x=top_countries.values, y=top_countries.index, palette=black_palette)
plt.title('Top Countries by Number of Sellers', fontsize=16)
plt.xlabel('Number of Sellers')
plt.ylabel('Country')
plt.show()
```

```python
# 5. Product Categories by Gender Target
plt.figure(figsize=(10, 6))
sns.countplot(y='product_category', hue='product_gender_target', data=df, palette=black_palette)
plt.title('Product Categories by Gender Target', fontsize=16)
plt.xlabel('Frequency')
plt.ylabel('Product Category')
plt.legend(title='Gender Target')
plt.show()
```

```python
# 6. Trend of Sold vs. Available Products
plt.figure(figsize=(10, 6))
sns.countplot(x='sold', data=df, palette=black_palette)
plt.title('Sold vs. Available Products', fontsize=16)
plt.xlabel('Sold')
plt.ylabel('Frequency')
plt.show()
```

```python
# Select only the numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = df[numeric_cols].corr()

# 7. Correlation Heatmap of Numerical Features
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='Greys', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```
